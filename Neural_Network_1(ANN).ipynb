# **How does a neural network work?**

Neural networks, inspired by the structure and function of the human brain, are powerful computational models employed in machine learning and deep learning. They are essentially systems of interconnected nodes (or neurons) organized into layers, trained to recognize patterns, classify data, and make predictions.

**Input Layer:** This layer receives the raw input data (e.g., pixels from an image, words from a sentence).

**Hidden Layers:** One or more layers situated between the input and output layers that perform the majority of the data processing and computations. Deep learning networks are characterized by having multiple hidden layers.

**Output Layer:** This final layer produces the network's prediction or classification (e.g., identifying the object in an image, predicting a stock price).



**why we assign weights to neural network input?**

Weights are assigned to neural network inputs to represent the strength or importance of the connection between that input and the neuron it feeds into. This concept is fundamental to how neural networks learn and make predictions.


**Epochs** In the context of neural networks and machine learning, an epoch represents one complete pass of the entire training dataset through the learning algorithm. During an epoch, every training example in the dataset is presented to the neural network, and the model's internal parameters (weights and biases) are updated based on the calculated error.


# Type of Loss

There are various loss functions used in neural networks, each suited for different types of problems. Here are some common ones:

**Mean Squared Error (MSE)**: Commonly used for regression tasks, it calculates the average of the squared differences between the predicted and actual values.

**Mean Absolute Error (MAE):** Also for regression, it calculates the average of the absolute differences between predicted and actual values. Less sensitive to outliers than MSE.

**Binary Crossentropy:** Used for binary classification problems (two classes), it measures the difference between the predicted probability and the true binary label.

**Categorical Crossentropy:** Used for multi-class classification (more than two classes) where each sample belongs to exactly one class. It compares the predicted class probabilities to the true class.

**Sparse Categorical Crossentropy:** Similar to categorical crossentropy, but used when the true labels are integers instead of one-hot encoded vectors. This is the one you used in the notebook.
The choice of loss function depends on the nature of the problem you are trying to solve (regression, binary classification, multi-class classification) and the characteristics of your data.

```python
model =keras.Sequential(
    [keras.layers.Dense(10,input_shape=(784,),activation='softmax')
    ])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'] #more than two classess
)
model.fit(x_train_flattened,y_train,epochs=5) # .fit is uded to train the model

```
